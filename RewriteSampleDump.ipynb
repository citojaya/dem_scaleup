```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os.path
import seaborn as sns
from scipy import stats
import math

%matplotlib inline


def readDump(fin, val):
    data = []
    with open (fin) as f:
        for line in f:
            if(line.strip() == "8.05E3"):
#                 print(line)
                data.append(val+"\n")
            elif(line.strip() == "8.05E3   1.0E7   2.9E-1  0.1   7.0E-5 0.1 1.1E-19"):
#                 print(line)
                data.append(val+"   1.0E7   2.9E-1  0.1   7.0E-5 0.1 1.1E-19\n")
            else:
                data.append(line)
    return data
                
def writeFile(fout, data):
    f = open(fout, 'w')
    for line in data:
#         print(line)
        f.write(line)
    f.close()

path_list = ["E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.3J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.3J/0.6N",
              "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.3J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.4J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.4J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.4J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.5J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.5J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/12.7mm/0.5J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.3J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.3J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.3J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.4J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.4J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.4J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.5J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.5J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/19.05mm/0.5J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.3J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.3J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.3J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.4J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.4J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.4J/0.7N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.5J/0.5N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.5J/0.6N",
             "E:/drum/density/2000/ppsf/0.1/damp/0.1/pwsf/0.4/254mm/25.4mm/0.5J/0.7N"]


out_data = []
for ls in path_list:
    fin = os.path.join(ls,'infile.in')
    data = readDump(fin, '2.05E3')

    fout = os.path.join(ls,'infile.in')
    writeFile(fout,data)

    
print('DONE')

```
